The following function is used to split and stack the nested values of the correponding columns 

```python
def preprocess(col):
    constraints = df_netflix[col].apply(lambda x : str(x).split(',')).tolist()
    df_of_col = pd.DataFrame(constraints, index = df_netflix['title'])    
    df = pd.DataFrame(df_of_col.stack()).reset_index()\
        .drop(columns = 'level_1').rename(columns = {0 : col})
    df[col] = df[col].str.strip()                             
    df[col] =  df[col].replace('nan', np.nan)
    return df
```

```python
#Bar plot
def plot_top_10(df, col):
    top_10 = df[col].value_counts().sort_values(ascending=False).head(10)
    plt.figure(figsize=(6,4))
    sns.barplot(x = top_10.index, y = top_10.values )
    plt.xticks(rotation = 70);
    plt.ylabel('Count of contents')
    plt.title(f'{col} with most no. of movies/TV shows');
```

```python
pd.merge(year_wise_movies_count, year_wise_shows_count, how = 'outer', left_index=True, right_index=True).\
sort_index(ascending=False)
```
